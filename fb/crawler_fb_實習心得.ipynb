```python
def get_date(str_date):
    date_obj = datetime.datetime.now()
    if("·" in str_date):
        str_date = str_date.split("·")[0]
    try:
        this_year = datetime.datetime.now().year
        if "分鐘" in str_date:
            date_obj  = datetime.datetime.now()
        if "小時" in str_date:
            date_obj  = datetime.datetime.now()
        if "昨天" in str_date:
            date_obj  = datetime.datetime.now()
        elif "年" in str_date:
            if "上午" in str_date:
                date_obj = datetime.datetime.strptime(str_date, "%Y年%m月%d日上午%H:%M")
            elif "下午" in str_date:
                date_obj = datetime.datetime.strptime(str_date, "%Y年%m月%d日下午%H:%M")
            else:
                date_obj = datetime.datetime.strptime(str_date, "%Y年%m月%d日")
        else:
            if "上午" in str_date:
                date_obj = datetime.datetime.strptime(str(this_year) + "年" + str_date, "%Y年%m月%d日上午%H:%M")
            elif "下午" in str_date:
                date_obj = datetime.datetime.strptime(str(this_year) + "年" + str_date, "%Y年%m月%d日下午%H:%M")
    except:
        print(str_date)
        date_obj  = datetime.datetime.now()
    return date_obj.strftime("%Y-%m-%d")
```

```python
def get_db_url(articles):
    result=[]
    count = 0
    for article in articles:
        data = {}
        link = article.select("footer > div:nth-child(2) > a")
        url = link[0]['href']
        date_str = article.select("footer > div:nth-child(1)")[0].text
        date = get_date(date_str)
        like_str = article.select("footer > div:nth-child(2) > span > a")[0].text
        like_str = re.findall(r'\d+', like_str)
        like_str = int(like_str[0]) if(len(like_str) > 0) else 0
        # 留言
        comment_num = article.select("footer > div:nth-child(2) > a")[0].text
        comment_num = re.findall(r'\d+', comment_num)
        c_num = int(comment_num[0]) if (len(comment_num) > 0) else 0
        data['Url'] = url
        data['post_date'] = date
        data['likes'] = like_str
        data['comment_num'] = c_num
        result.append(data)
    return result
```

```python
def getComment(soup):
    commentList = soup.find("div", id = "m_story_permalink_view").contents[-1].contents[0].contents[-3].contents
    commten_list = []
    if(len(commentList)>0 and commentList[0].name=='div'):
        for comment in commentList:
            if(len(comment.find_all("div"))>0):
                data = {}
                poster_a = comment.find("a")
                comment_content = comment.find_all("div")[1].text
                data['poster'] = poster_a.text
                data['posterUrl'] = url+poster_a['href']
                data['comment'] = comment_content
                commten_list.append(data)
    if(len(commten_list)>0):
        return commten_list
    else:
        return None
```

```python
def findInfo(Url_dict):
    art_dict = {}
    try:
       driver.get(Url_dict['Url'])
    except Exception as e:
        print(e)
        art_dict['content'] = []
        return art_dict
        # return
    resp = driver.page_source
    resp = resp.replace("<br>", "。")
    soup = BeautifulSoup(resp, "html.parser")
    
    try:
        srt_date = soup.find("footer").find("abbr").text
    except:
        srt_date=''
    if(srt_date == ""): 
        art_dict['content'] = []
        return art_dict
    this_year = datetime.datetime.now().year
    if "小時" in srt_date:
        date_obj  = datetime.datetime.now()
    if "昨天" in srt_date:
        date_obj  = datetime.datetime.now()
    elif "年" in srt_date:
        if "上午" in srt_date:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日上午%H:%M")
        elif "下午" in srt_date:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日下午%H:%M")
        else:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日")
    else:
        if "上午" in srt_date:
            date_obj = datetime.datetime.strptime(str(this_year) + "年" + srt_date, "%Y年%m月%d日上午%H:%M")
        elif "下午" in srt_date:
            date_obj = datetime.datetime.strptime(str(this_year) + "年" + srt_date, "%Y年%m月%d日下午%H:%M")

    art_dict['date'] = date_obj.strftime("%Y-%m-%d")
    # 內容
    texts_html = soup.find("div", id = "m_story_permalink_view").select_one('div > div').select('div > div')[1].select('div > div')[1].text
    art_dict['Url'] = Url_dict['Url']
    art_dict['content'] = texts_html
    art_dict['reaction_like_count'] = Url_dict['likes']
    art_dict['comment_num'] = Url_dict['comment_num']
    poster = soup.find("header").find("header").find("h3").find("a")
    art_dict['poster'] = poster.text
    art_dict['posterUrl'] = url+poster['href']
    art_dict['comments'] = getComment(soup)

    return art_dict
```
