```python
# 不同公司
for c in company_code:
    try:
        # 不同頁面
        time.sleep(2)
        url = f'https://interview.tw/c/{c}?sort=newest'
        driver.get(url)
        resp = driver.page_source
        soup = BeautifulSoup(resp,"html.parser")
        
        # 公司名稱
        _company_name = soup.find('h1', {'class': 'fz-h2 fz-tit'}).text
        _location = soup.find('div', {'class': 'iw_info-list'}).find('div', {'class': 'iw_city-item'}).find('span', {'class': 'fz-s3'}).text
        iw_li_list = soup.find('ul', {'class': 'iw_breadcrumb-wrap'}).find_all('li')
        industry_big = iw_li_list[1].find('a').text
        if(industry_big!='其他'):
            industry_mid = iw_li_list[2].find('a').text
            industry_li = iw_li_list[3].find('a').text
        else:
            industry_mid = ''
            industry_li = ''
        info_list = soup.find('div', {'class': 'iw_company-info-list'}).find_all('div', {'class':'iw_main-info'})
        info_list = [i.find("span").text for i in info_list]
        _grade = info_list[0]
        _degree = info_list[1]
        _salary_m = info_list[2]
        _postNum = info_list[3]
        _infoNum = info_list[4]
        tmp_df_interview = pd.DataFrame({
            "公司名稱": [_company_name],
            "company_code":[str(c)],
            "產業別(大類)":[industry_big],
            "產業別(中類)":[industry_mid],
            "產業別(小類)":[industry_li],
            "地區":[_location],
            "面試評價":[_grade],
            "面試難度":[_degree],
            "平均月薪":[_salary_m],
            "面試心得數":[_postNum],
            "薪資情報數":[_infoNum]
        }, index=[len(df_interview)])
        df_interview = df_interview.append(tmp_df_interview)
    except:
        driver.refresh()
```
