### 爬取文章連結、按讚數、留言數

```python
# link_list = []
count_list =[]
result = []
for i in range(500):
    try:
        resp = driver.page_source
        soup = BeautifulSoup(resp,"html.parser")
        articlesRoot = soup.find_all('div', id='m_group_stories_container') #文章最上層
        articles = (articlesRoot[0].find_all("article", class_ = "da dc dm")) #articles per page
        for article in articles:
            result_dic = {}
            try:
                link = article.select(".dn .ds a:not([class])")[0]['href']
                likes = article.find("footer").find_all("div", class_ = "cp cq")[-1:][0].find("a", class_ = "ef eg").text
                comment_num = article.find("footer").find_all("div", class_ = "cp cq")[-1:][0].find_all("a", class_ = "eg")[-1:][0].text
                comment_list = re.findall(r'\d+', comment_num)
                count = 0
                if(len(comment_list)==1):
                    count = int(comment_list[0])
                count
            except Exception as e:
                print(e)
                likes = 0
                count = 0
                link = ''
            result_dic['artUrl'] = link
            result_dic['likes'] = likes
            result_dic['comment_num'] = count
            if("https://mbasic.facebook.com/" in link):
                result.append(result_dic)
        count_list.append(len(articles))
        driver.find_element(By.PARTIAL_LINK_TEXT,"查看更多貼文").click()
    except Exception as e:
        print(e)
```

```python
def findInfo(Url_dict):
    driver.get(Url_dict['artUrl'])
    resp = driver.page_source
    resp = resp.replace("<br>", "。")
    soup = BeautifulSoup(resp, "html.parser")
    art_dict = {}
    try:
        srt_date = soup.find("footer").find("abbr").text
    except:
        srt_date=''
    if(srt_date == ""): 
        art_dict['content'] = []
        return art_dict
    this_year = datetime.datetime.now().year
    if "小時" in srt_date:
        date_obj  = datetime.datetime.now()
    if "昨天" in srt_date:
        date_obj  = datetime.datetime.now()
    elif "年" in srt_date:
        if "上午" in srt_date:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日上午%H:%M")
        elif "下午" in srt_date:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日下午%H:%M")
        else:
            date_obj = datetime.datetime.strptime(srt_date, "%Y年%m月%d日")
    else:
        if "上午" in srt_date:
            date_obj = datetime.datetime.strptime(str(this_year) + "年" + srt_date, "%Y年%m月%d日上午%H:%M")
        elif "下午" in srt_date:
            date_obj = datetime.datetime.strptime(str(this_year) + "年" + srt_date, "%Y年%m月%d日下午%H:%M")
    
    # comment_num = 0
    # tmp = soup.find_all("div", class_ = ["dq","du","dw","ds","dr","dp","dm","dk","df","di"])
    # tmplist = [i.text for i in tmp]
    # regex = re.compile(r'^\d+$')
    # matched_items = [item for item in tmplist if regex.match(item)]
    # if(len(matched_items) == 0):
    #     print("run")
    #     for s in tmplist:
    #         pattern = r"其他\s+(\d+)"
    #         match = re.search(pattern, s)
    #         if match:
    #             matched_items.append(match.group(1))
    #             break
    # if(len(matched_items) > 0):
    #     comment_num = matched_items[0]
    # int(comment_num)

    art_dict['date'] = date_obj.strftime("%Y-%m-%d")
    # 內容
    texts_html = soup.find("div", class_ = "bi").find("div", class_ = "bj bk").find("div", class_ = "bl").find("div", class_ = ["bu", "bw", "by","ca", "bv"]).text
    art_dict['artUrl'] = Url_dict['artUrl']
    art_dict['content'] = texts_html
    art_dict['reaction_like_count'] = Url_dict['likes']
    art_dict['comment_num'] = Url_dict['comment_num']
    poster = soup.find("header").find("header").find("h3").find("a")
    art_dict['poster'] = poster.text
    art_dict['posterUrl'] = poster['href']

    return art_dict
```
